```python
# 🛍️ Customer Shopping Behavior Analysis

## Objective

The objective of this project is to analyze customer shopping behavior using Python, Microsoft SQL Server, and Power BI. The project follows an end-to-end data analytics workflow including data cleaning, exploratory data analysis (EDA), SQL-based business analysis, and dashboard creation to generate actionable business insights.

### Tools Used

- Python
- Pandas
- NumPy
- Matplotlib
- Microsoft SQL Server
- Power BI
```

```python

# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported successfully!")
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
print("Dataset loaded successfully!")
```

```python
df.head()
```

```python
# Check the number of rows and columns

df.shape
```

```python
df.info()
```

```python
# checking the stastical summary of the data 
df.describe(include='all')
```

```python

# Check Missing Values
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
# Export cleaned dataset
df.to_csv("customer_shopping_cleaned.csv", index=False)

print("Dataset exported successfully!")
```
